```python
filename = "/home/fhr2592/workspace/hw-physics390/cpp/probability/data/uniform.dat"

def show_file(filename):
    values = []

    with open(filename, "r") as f:
        for line in f:
            if "val: " in line:
                value = float(line.split("val: ")[1].strip())
                values.append(value)

    plt.hist(values, bins=100, edgecolor='black')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    # plt.show()

    print("N ",len(values))
    print("mean", np.mean(values), "std",  np.std(values))
    print(values[:3])

show_file(filename)
plt.show()
```

```python
from math import pi, sqrt

a=1

def show_mb_results(a):
    print(f"Showing MB results for {a=}")

    filename = f'/home/fhr2592/workspace/hw-physics390/cpp/probability/data/maxwell_boltzmann_rejection_a{a}.dat'
    show_file(filename)

    def get_mb_expected_stats(a):
        """Calculate expected statistics from a boltzman distribution (see https://en.wikipedia.org/wiki/Maxwell%E2%80%93Boltzmann_distribution#Distribution_function)"""
        mean = 2*a*sqrt(2/pi)
        var = a**2 * (3*pi-8) / pi
        std = sqrt(var)
        return mean, std

    expected_mean, expected_std = get_mb_expected_stats(a)
    print(f"{expected_mean=}", f"{expected_std=}", sep='\n')

for a in [1,2,5]:
    show_mb_results(a)

plt.show()
```
